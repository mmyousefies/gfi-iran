```plaintext
library(tidyverse)
```

```plaintext
notorious.tbl <- tibble(weight = runif(2000,1e-5,2),weight2= abs(rnorm(2000,1,1)),age = rnorm(2000,50,1))
notorious.tbl|> 
  ggplot(aes(x = age, weight = weight))+
  geom_density(color = 'black')
```

```plaintext
expanded.tbl <- notorious.tbl |> 
  mutate(weight_s = weight/min(weight)) |> 
  rowwise() |> 
  mutate(reps = round(weight_s*10)) |> 
  uncount(reps)
expanded.tbl |> ggplot(aes(x = age)) +geom_density(color = 'black')
```
